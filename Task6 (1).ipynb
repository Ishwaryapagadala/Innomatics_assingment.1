```python
def parking(capacity,logs):
  count= 0
  for i in logs:
    if i =="IN":
      count +=1
    else:
      count -=1
  print("Currently Parked Vechicles:",count)
  print("Parking Status:","Full" if count >= capacity else "Available")
parking(50,["IN","IN","IN","OUT","IN","IN","OUT"])
```

```python
def delivery(distance,traffic,weather):
  time = distance *5
  if traffic == "High":
    time +=10
  if weather == "Rainy":
    time+=5
  print("Estimated Delivery Time:",time,"minutes")
delivery(8,"High","Rainy")
```

```python
def seats(total,booked):
  percent = len(booked)* 100/total
  if percent >=75:
    print("Occupancy:",int(percent),"%")
    print("Show Status:Almost Full")
seats(200,[1]*150)
```

```python
def load(cpu):
  avg = sum(cpu)/len(cpu)
  if avg <= 80:
    print("Average CPU Load:",int(avg),"%")
    print("Server Status:Warning")
load([45,60,70,85,90])
```

```python
def resource(d):
  for i in d:
    if d[i]>8:print(i,"Yes")
resource({"Project":6,"AC":9,"Lights":4})
```
