```python
def analyze_engagement(likes):
    total = sum(likes)
    print("Total Likes:",total)
    print("Post Status:","Viral Post" if total >= 1000 else "Normal Engagement")
analyze_engagement([200,300,150,400])
```

```python
def stock_alert(stock):
  print("Medicine Stock:",stock)
  print("Status:","Low Stock Alert" if stock < 10 else "stock sufficient")
stock_alert(6)
```

```python
def rainfall_status(data,required):
  avg = sum(data) /len(data)
  print("Average Rainfall:",avg)
  print("Rainfall Status:","Adequate Rainfall" if avg >= required else "Inadeuate Rainfall")
rainfall_status([70,75,80,65],70)
```

```python
def detect_duplicates(users):
  print("Duplicate Accounts Found:",
        "Yes" if len(users) != len(set(users)) else "No")
detect_duplicates(["john","alice","john","mark"])
```

```python
def check_eligibility(age):
  print("Patient Age:",age)
  print("Eligibility Status:","Eligibility" if age >=18 else "Not Eligible")
check_eligibility(21)
```

```python
def premium_crops(prices):
  print("Premium Crops:",[p for p in prices if p > 20000])
premium_crops([1500,2500,3200,1800])
```

```python
def system_health(errors):
  status = "Healthy" if errors ==0 else \
            "Minor Issues" if errors <=5 else "Critical Issues"
  print("Error Count:",errors)
  print("System Status",status)
system_health(7)
```

```python
def check_transaction(amount):
  print("Transaction Amount:",amount)
  print("Transaction" if amount <= 50000 else "Rejected")
check_transaction(60000)
```

```python
def attendence_check(records):
  percent= records.count("Present") / len(records)*100
  print("Attendence Percentage:",percent)
  print("Exam Eligibility:","Eligibility" if percent >= 75 else "Not Eligible")
attendence_check(["Present","Present","Absent","Present","Present"])
```

```python
def electricity_bill(u):
  b= u*3 if u <= 100 else u*5-200 if u<=200 else u*7-600
  print("Total Bill:",b,"\nUsage Status:","Low Usage" if b<500 else"High Usage")
electricity_bill(250)
```
