```python
employees = {"Ravi":92,"Anita":88,"Kiran":92,"Suresh":85}
m = max(employees.values())
print("Top performers Eligible for Bonus:",
      ", ".join([k for k, v in employees.items() if v==m]),
      f"(Score: {m})")
```

```python
from collections import Counter
q = "Buy mobile phone buy phone online".lower().split()
print({k:v for k,v in Counter(q).items() if v > 1})
```

```python
s = [3,4,7,8,10,12,5]
print([(i,v) for i,v in enumerate(s) if v % 2 ==0])
```

```python
from collections import Counter
emails =["ravi@gmail.com","anita@yahoo.com","kiran@gmail.com","suresh@gmail.com"]
c = Counter(e.split("@")[1] for e in emails)
for k,v in c.items():
  print(f"{k}:{v/len(emails):.0%}")
```

```python
sales = [1200,1500,900,2200,1400,3000]
avg = sum(sales)/len(sales)
for i,v in enumerate(sales,1):
  if v> avg*1.3:
    print(f"Day {i}:{v}")
```
