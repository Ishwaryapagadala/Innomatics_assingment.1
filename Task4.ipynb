```python
def recharge():
  while (a:=int(input("Amount: "))) not in [199,299,399,599] or a <50:
    print("Invalid, Try Again")
  print("Recharge Successful")
recharge()
```

```python
def alert(d):
  for k,v in d.items(): print(k,":","Reorder" if v<15 else "OK")
alert({"Pen":10,"Book":20,"Bag":5})
```

```python
def cab():
  while 1:
    f=50+float(input("km: "))*12
    if input("Peak(y/n): ")=="y": f*=1.25
    print("Fare:",f)
    if input("Retry(y/n): ")!="y": break
cab()
```
